```python
# Define and retrieve parameters
dbutils.widgets.text("catalog_name", "lending_risk_dev") # Default value for manual testing
CATALOG = dbutils.widgets.get("catalog_name")

print(f"Running setup for Catalog: {CATALOG}")
```

```python
# Create Payment Performance Table in Gold Layer
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {CATALOG}.gold.calc_payment_performance 
(
  member_id                 STRING NOT NULL,
  last_payment_pts          INT,
  total_payment_pts         INT,    
  _updated_at               TIMESTAMP,
  _batch_id                 STRING,      
  _scoring_module           STRING,
  
  CONSTRAINT payment_perf_pk PRIMARY KEY (member_id)
)
USING DELTA
CLUSTER BY (member_id) 
TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true',
    'delta.autoOptimize.autoCompact' = 'true',
    'delta.enableDeletionVectors' = 'true',
    'delta.columnMapping.mode' = 'name',
    'delta.minReaderVersion' = '3',
    'delta.minWriterVersion' = '7'
)
""")
```

```python
# Create Defaulters Performance Table in Gold Layer
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {CATALOG}.gold.calc_defaulters_performance 
(
  member_id                 STRING NOT NULL,
  delinq_pts                INT,
  public_records_pts        INT,
  public_bankruptcies_pts   INT,
  inquiry_pts               INT, 
  _updated_at               TIMESTAMP,
  _batch_id                 STRING,      -- Added for Execution Traceability
  _scoring_module           STRING,
  
  -- Inform the Optimizer that uniqueness is guaranteed by the pipeline
  CONSTRAINT defaulter_perf_pk PRIMARY KEY (member_id) RELY
)
USING DELTA
CLUSTER BY (member_id) 
TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true',
    'delta.autoOptimize.autoCompact' = 'true',
    'delta.enableDeletionVectors' = 'true',
    'delta.columnMapping.mode' = 'name',
    'delta.minReaderVersion' = '3',
    'delta.minWriterVersion' = '7'
)
""")
```

```python
# Create Financial Performance Table in Gold Layer
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {CATALOG}.gold.calc_financial_performance 
(
  member_id                 STRING NOT NULL,
  loan_status_pts           INT,
  home_pts                  INT,
  credit_limit_pts          INT,
  grade_pts                 INT,  
  _updated_at               TIMESTAMP,
  _batch_id                 STRING,       
  _scoring_module           STRING,
  
  -- Inform the Optimizer that uniqueness is guaranteed by the pipeline
  CONSTRAINT fin_perf_pk PRIMARY KEY (member_id) RELY
)
USING DELTA
CLUSTER BY (member_id)
TBLPROPERTIES (
    'delta.enableChangeDataFeed' = 'true',
    'delta.autoOptimize.optimizeWrite' = 'true',
    'delta.autoOptimize.autoCompact' = 'true',
    'delta.enableDeletionVectors' = 'true',
    'delta.columnMapping.mode' = 'name',
    'delta.minReaderVersion' = '3',
    'delta.minWriterVersion' = '7'
)
""")
```
