```python
def generate_merge_query(target_table: str, source_view: str, merge_condition: str, set_col_ls: list, insert_col_ls: list) -> str:
    """
    Generates a dynamic SQL MERGE statement based on provided column lists and conditions.
    """
    # 1. Build the UPDATE SET clause mappings
    set_statements = [f"tgt.{col} = src.{col}" for col in set_col_ls]
    set_clause = ", ".join(set_statements)
    
    # 2. Build the INSERT columns and values
    insert_cols = ", ".join(insert_col_ls)
    insert_vals = ", ".join([f"src.{col}" for col in insert_col_ls])
    
    # 3. Construct the final query string
    query = f"""
    MERGE INTO {target_table} AS tgt
    USING {source_view} AS src
    ON {merge_condition}
    WHEN MATCHED THEN
      UPDATE SET {set_clause}
    WHEN NOT MATCHED THEN
      INSERT ({insert_cols})
      VALUES ({insert_vals})
    """
    return query.strip()
```
